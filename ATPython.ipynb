4) A partir do dicionário alterado na questão 2, busque por todos os alunos que têm seu nome iniciado pela letra "G" e elimine-os do dicionário. Armazene este novo dicionário em um arquivo texto chamado "novo_arquivo.txt".

```python
import requests

url = "https://raw.githubusercontent.com/cassiusf/datasets/main/alunos_nomes.txt"
arquivo = requests.get(url)
conteudo = arquivo.text

alunos_dict = {}
for linha in conteudo.splitlines():
    key, value = linha.split(' - ')
    alunos_dict[int(key)] = value

alunos_dict = {key: value.lstrip('0123456789-').strip() for key, value in alunos_dict.items()}

alunos_a_remover = [key for key, value in alunos_dict.items() if value.upper().startswith('G')]
for key in alunos_a_remover:
    del alunos_dict[key]

with open('novo_arquivo.txt', 'w') as file:
    for key, value in alunos_dict.items():
        file.write(f"{key} - {value}\n")

print("Novo dicionário salvo em 'novo_arquivo.txt' com sucesso!")

with open('novo_arquivo.txt', 'r') as file:
    novo_conteudo = file.read()

print("\nConteúdo do arquivo 'novo_arquivo.txt':")
print(novo_conteudo)
```

10) A partir da base carregada na questão 9, apresente as médias de preço e quilates ("carat") e a mediana da variável "depth" para os diamantes da base, agrupando-os pelas variáveis "cut" e "color".

```python
import pandas as pd

url = "https://github.com/cassiusf/datasets/raw/main/diamonds.xlsx"
df = pd.read_excel(url, engine='openpyxl')

agrupado = df.groupby(['cut', 'color'])

agregado = agrupado.agg({
    'price': 'mean',
    'carat': 'mean',
    'depth': 'median'
}).reset_index()

agregado = agregado.rename(columns={
    'price': 'average_price',
    'carat': 'average_carat',
    'depth': 'median_depth'
})

print(agregado)
```

11) Carregue a base em SQLite que está em “https://github.com/cassiusf/datasets/blob/main/mpg.db” para um dataframe. Faça as conversões de tipo necessárias para ajustar as variáveis aos dados de forma correta.


```python
import sqlite3
import pandas as pd
import requests

url = "https://github.com/cassiusf/datasets/raw/main/mpg.db"

db_filename = "mpg.db"

response = requests.get(url)
with open(db_filename, 'wb') as f:
    f.write(response.content)

conn = sqlite3.connect(db_filename)

names_df = pd.read_sql_query("SELECT * FROM Name;", conn)
observations_df = pd.read_sql_query("SELECT * FROM Observation;", conn)
origins_df = pd.read_sql_query("SELECT * FROM Origin;", conn)

conn.close()

observations_df['mpg'] = observations_df['mpg'].astype(float)
observations_df['cylinders'] = observations_df['cylinders'].astype(int)
observations_df['displacement'] = observations_df['displacement'].astype(float)
observations_df['horsepower'] = observations_df['horsepower'].replace('?', pd.NA).astype(float)
observations_df['weight'] = observations_df['weight'].astype(float)
observations_df['acceleration'] = observations_df['acceleration'].astype(float)
observations_df['model_year'] = observations_df['model_year'].astype(int)
observations_df['origin_id'] = observations_df['origin_id'].astype(int)
observations_df['name_id'] = observations_df['name_id'].astype(int)

merged_df = observations_df.merge(names_df, on='name_id').merge(origins_df, on='origin_id')

print(merged_df.head())
```

12) A partir da base carregada na questão 11, apresente as médias de consumo ("mpg"), potência ("horsepower") e peso ("weight") por origem e quantidade de cilindros. Na mesma tabela final, apresente também a quantidade de carros para cada ano existente na base, a partir da variável "model_year".

```python
import sqlite3
import pandas as pd
import requests

url = "https://github.com/cassiusf/datasets/raw/main/mpg.db"

db_filename = "mpg.db"

response = requests.get(url)
with open(db_filename, 'wb') as f:
    f.write(response.content)

conn = sqlite3.connect(db_filename)

names_df = pd.read_sql_query("SELECT * FROM Name;", conn)
observations_df = pd.read_sql_query("SELECT * FROM Observation;", conn)
origins_df = pd.read_sql_query("SELECT * FROM Origin;", conn)

conn.close()

observations_df['mpg'] = observations_df['mpg'].astype(float)
observations_df['cylinders'] = observations_df['cylinders'].astype(int)
observations_df['displacement'] = observations_df['displacement'].astype(float)
observations_df['horsepower'] = observations_df['horsepower'].replace('?', pd.NA).astype(float)
observations_df['weight'] = observations_df['weight'].astype(float)
observations_df['acceleration'] = observations_df['acceleration'].astype(float)
observations_df['model_year'] = observations_df['model_year'].astype(int)
observations_df['origin_id'] = observations_df['origin_id'].astype(int)
observations_df['name_id'] = observations_df['name_id'].astype(int)

# Calcular médias de consumo, potência e peso por origem e quantidade de cilindros
means_df = observations_df.groupby(['origin_id', 'cylinders']).agg({'mpg': 'mean', 'horsepower': 'mean', 'weight': 'mean'}).reset_index()

# Calcular a quantidade de carros para cada ano de modelo
car_count_df = observations_df.groupby('model_year').size().reset_index(name='car_count')

# Renomeie a coluna 'model_year' em car_count_df para 'origin' para que corresponda à coluna em means_df
car_count_df.rename(columns={'model_year': 'origin'}, inplace=True)

# Mesclar as duas tabelas com base na coluna 'origin'
merged_df = pd.concat([means_df, car_count_df], axis=1)

print("Colunas em means_df:", means_df.columns)
print("Colunas em car_count_df:", car_count_df.columns)

print(merged_df)
```

13) Faça um código em Python que tente carregar a base disponível em “https://github.com/cassiusf/datasets/blob/main/films.json” para um dataframe. Implemente uma estrutura completa de tratamento de erros, utilizando a estrutura "try...except...else...finally" que verifique se o arquivo está acessível e, caso não esteja, emita uma mensagem informando o usuário do problema. Não esqueça de implementar de forma coerente as estruturas "else..." e "finally...".

```python
import pandas as pd
import requests
import io

url = "https://github.com/cassiusf/datasets/raw/main/films.json"

try:
    response = requests.get(url)
    response.raise_for_status()  # Raises an exception for 4xx or 5xx status codes
except requests.exceptions.RequestException as e:
    print("Erro ao acessar o arquivo:", e)
else:
    try:
        df = pd.read_json(io.BytesIO(response.content))
        print("Base de dados carregada com sucesso:")
        print(df.head())
    except ValueError as e:
        print("Erro ao decodificar o JSON:", e)
finally:
    print("Finalizando o processo de carregamento.")
```
